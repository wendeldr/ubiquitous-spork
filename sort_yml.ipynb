```python
# read yaml file and extract labels tag
import yaml
import copy
import pandas as pd

with open('used.yaml', 'r') as file:
    data = yaml.safe_load(file)

print(data)
```

```python
labels = []
for key, value in data.items():
    for k, v in value.items():
        labels.extend(v['labels'])

labels = set(labels)
# print(labels)
```

```python
cols = ['measure','unsigned', 'linear', 'signed', 'undirected', 'contemporaneous', 'nonlinear', 'frequency-dependent', 'unordered', 'multivariate', 'bivariate', 'directed', 'time-dependent']
dict_data = {col: 0 for col in cols}
rows = []
for key, value in data.items():
    for k, v in value.items():
        tmp = copy.deepcopy(dict_data)
        tmp['measure'] = k
        for label in v['labels']:
            if label == "signed/unsigned":
                tmp['signed'] = 1
                tmp['unsigned'] = 1
            elif label == "linear/nonlinear":
                tmp['nonlinear'] = 1
                tmp['linear'] = 1
            else:
                tmp[label] = 1
        rows.append(tmp)

df = pd.DataFrame(rows)

```
